```python
import pandas as pd
AAPL = pd.read_csv('AAPL.csv') 
AAPL_profit = []
AAPL_profit.append(0)
base_day = -500
for i in range(500):
    current = int(i + -500)
    AAPL_profit.append((AAPL.iloc[current].Close/AAPL.iloc[base_day].Close) - 1)
    
MSFT = pd.read_csv('MSFT.csv') 
MSFT_profit = []
MSFT_profit.append(0)
base_day = -500
for i in range(500):
    current = int(i + -500)
    MSFT_profit.append((MSFT.iloc[current].Close/MSFT.iloc[base_day].Close) - 1)
    
MMM = pd.read_csv('MMM.csv') 
MMM_profit = []
MMM_profit.append(0)
base_day = -500
for i in range(500):
    current = int(i + -500)
    MMM_profit.append((MMM.iloc[current].Close/MMM.iloc[base_day].Close) - 1)
    
AMZN = pd.read_csv('AMZN.csv') 
AMZN_profit = []
AMZN_profit.append(0)
base_day = -500
for i in range(500):
    current = int(i + -500)
    AMZN_profit.append((AMZN.iloc[current].Close/AMZN.iloc[base_day].Close) - 1)
    
GOOG = pd.read_csv('GOOG.csv') 
GOOG_profit = []
GOOG_profit.append(0)
base_day = -500
for i in range(500):
    current = int(i + -500)
    GOOG_profit.append((GOOG.iloc[current].Close/GOOG.iloc[base_day].Close) - 1)
    
JPM = pd.read_csv('JPM.csv') 
JPM_profit = []
JPM_profit.append(0)
base_day = -500
for i in range(500):
    current = int(i + -500)
    JPM_profit.append((JPM.iloc[current].Close/JPM.iloc[base_day].Close) - 1)
    
BBT = pd.read_csv('BBT.csv') 
BBT_profit = []
BBT_profit.append(0)
base_day = -500
for i in range(500):
    current = int(i + -500)
    BBT_profit.append((BBT.iloc[current].Close/BBT.iloc[base_day].Close) - 1)
    
S_P_500 = pd.read_csv('^GSPC.csv') 
S_P_500_profits = []
S_P_500_profits.append(0)
base_day = -500
for i in range(500):
    current = int(i + -500)
    S_P_500_profits.append((S_P_500.iloc[current].Close/S_P_500.iloc[base_day].Close) - 1)
```

```python
plt.figure(figsize=(15,6))
plt.plot(AAPL_profit)
plt.plot(MSFT_profit)
plt.plot(MMM_profit)
plt.plot(AMZN_profit)
plt.plot(GOOG_profit)
plt.plot(JPM_profit)
plt.plot(BBT_profit)
plt.plot(S_P_500_profits)
plt.legend(['AAPL', 'MSFT', 'MMM', 'AMZN', 'GOOG', 'JPM', 'BBT', 'S&P'])
```
